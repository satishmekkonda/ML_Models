```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200, n_features=2, centers=4,
                 cluster_std=1.8, random_state=101)
```

```python
data
```

```python
print(data[0])
print('--------')
print('\n')
print(data[1])
```

```python
data[0].shape
```

```python
data[1].shape
```

```python
plt.scatter(data[0][:, 0], data[0][:, 1], c=data[1], cmap='rainbow')
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters = 4)
kmeans.fit(data[0])


kmeans1 = KMeans(n_clusters = 6)
kmeans1.fit(data[0])

kmeans2 = KMeans(n_clusters = 2)
kmeans2.fit(data[0])
```

```python
kmeans2.cluster_centers_
```

```python
kmeans2.labels_
```

```python
fig, (ax1, ax2, ax3, ax4, ax5, ax6) = plt.subplots(1, 6, sharey=True, figsize=(10, 6))

ax1.scatter(data[0][:, 0], data[0][:, 1], c=kmeans.labels_, cmap = 'rainbow')
ax2.scatter(data[0][:, 0], data[0][:, 1], c=data[1], cmap = 'rainbow')
ax1.set_title('K Means')

ax3.scatter(data[0][:, 0], data[0][:, 1], c=kmeans1.labels_, cmap = 'rainbow')
ax4.scatter(data[0][:, 0], data[0][:, 1], c=data[1], cmap = 'rainbow')
ax5.set_title('K Means')

ax5.scatter(data[0][:, 0], data[0][:, 1], c=kmeans2.labels_, cmap = 'rainbow')
ax6.scatter(data[0][:, 0], data[0][:, 1], c=data[1], cmap = 'rainbow')
ax5.set_title('K Means')
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(10, 6))

ax1.scatter(data[0][:, 0], data[0][:, 1], c=kmeans.labels_, cmap = 'rainbow')
ax2.scatter(data[0][:, 0], data[0][:, 1], c=data[1], cmap = 'rainbow')
ax1.set_title('K Means')

fig, (ax3, ax4) = plt.subplots(1, 2, sharey=True, figsize=(10, 6))
ax3.scatter(data[0][:, 0], data[0][:, 1], c=kmeans1.labels_, cmap = 'rainbow')
ax4.scatter(data[0][:, 0], data[0][:, 1], c=data[1], cmap = 'rainbow')
ax3.set_title('K Means')


fig, (ax5, ax6) = plt.subplots(1, 2, sharey=True, figsize=(10, 6))
ax5.scatter(data[0][:, 0], data[0][:, 1], c=kmeans2.labels_, cmap = 'rainbow')
ax6.scatter(data[0][:, 0], data[0][:, 1], c=data[1], cmap = 'rainbow')
ax5.set_title('K Means')
```
